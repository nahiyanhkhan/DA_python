# Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
df_used_cars
```

```python
plt.figure(figsize = (10,6))
plt.bar(x = df_used_cars['Brand'], 
        height = df_used_cars['Cars Listings'], 
        color = ['midnightblue'])

plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)

plt.title("Cars List by Brand", fontsize = 16, fontweight = 'bold')
plt.ylabel("Number of Listings\n", fontsize = 13)

plt.savefig("Used_Cars_Bar.png", bbox_inches = 'tight')
plt.show()
```

# Pie Chart

```python
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")
df_fuel_engine_type
```

```python
df_fuel_engine_type['Engine Fuel Type'].values
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10,8))

plt.pie(df_fuel_engine_type['Number of Cars'],
        labels = df_fuel_engine_type['Engine Fuel Type'].values,
        autopct = '%.2f%%',
        textprops = {'size':'x-large', 
                     'weight':'bold', 
                     'rotation':'38', 
                     'color':'w'})

plt.legend()
plt.title("Cars by Engine Fuel Type", fontsize = 16, fontweight = 'bold')

plt.show()
```

# Stacked Area Chart

```python
df_fuel_engine_types = pd.read_csv("stacked_area_chart_data.csv")
df_fuel_engine_types
```

```python
colors = ['#011638', '#7e2987', '#ef2026']
labels = ['Diesel', 'Petrol', 'Gas']
sns.set_style("white")

plt.figure(figsize = (12,6))

plt.stackplot(df_fuel_engine_types['Year'],
              df_fuel_engine_types['Diesel'],
              df_fuel_engine_types['Petrol'],
              df_fuel_engine_types['Gas'],
              colors = colors, edgecolor = 'None')

plt.xticks(df_fuel_engine_types['Year'], rotation = 50)
plt.legend(labels = labels, loc = 'upper left')

plt.ylabel("Number of Cars", fontsize = 14)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", 
          fontsize = 16, weight = 'bold')
sns.despine()

plt.savefig("Stacked_Area_Chart_by_Fuel_Type.png")

plt.show()
```

# Line Chart

```python
df_spx_ftse_00_10 = pd.read_csv("line_chart_data.csv")
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10['new_date'] = pd.to_datetime(df_spx_ftse_00_10['Date'])
df_spx_ftse_00_10['new_date']
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize = (20,8))

plt.plot(df_spx_ftse_00_10['new_date'], df_spx_ftse_00_10['GSPC500'])
plt.plot(df_spx_ftse_00_10['new_date'], df_spx_ftse_00_10['FTSE100'])

plt.xlabel("Date")
plt.ylabel("Returns")

plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize = 17, weight = 'bold')
plt.legend(labels = labels, fontsize = 'x-large')

plt.show()
```

```python
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >= '2008-07-01') & 
                                      (df_spx_ftse_00_10.new_date <= '2008-12-31')]
df_spx_ftse_H2_08
```

```python
# labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize = (20,8))

plt.plot(df_spx_ftse_H2_08['new_date'], df_spx_ftse_H2_08['GSPC500'], color = 'midnightblue')
plt.plot(df_spx_ftse_H2_08['new_date'], df_spx_ftse_H2_08['FTSE100'], color = 'crimson')

plt.xlabel("Date")
plt.ylabel("Returns")

plt.title("S&P vs FTSE Returns (H2 2008)", fontsize = 17, weight = 'bold')
plt.legend(labels = labels, fontsize = 'x-large')

plt.show()
```

# Histogram

# Scatter Plot

```python
df_real_estate = pd.read_csv("scatter_data.csv")
df_real_estate
```

```python
sns.set()
```

<b>plt.scatter()<b>

```python
plt.figure(figsize = (12,8))

scatter = plt.scatter(df_real_estate['Area (ft.)'],
                      df_real_estate['Price'],
                      alpha = 0.6,
                      c = df_real_estate['Building Type'],
                      cmap = 'viridis')

plt.legend(*scatter.legend_elements(),
           loc = 'upper left',
           title = "Building Type")
plt.title("Relationship between Area and Price in California Real Estate",
          fontsize = 15,
          weight = 'bold')
plt.xlabel("Area (sq. ft.)", weight = 'bold')
plt.ylabel("Price (000's $)", weight = 'bold')

plt.show()
```

<b>sns.scatterplot()<b>

```python
plt.figure(figsize = (12,8))

sns.scatterplot(x = df_real_estate['Area (ft.)'],
                y = df_real_estate['Price'],
                hue = df_real_estate['Building Type'],
                palette = ['black', 'darkblue', 'purple', 'pink', 'white'],
                s = 100)

plt.title("Relationship between Area and Price in California Real Estate",
          fontsize = 15,
          weight = 'bold')
plt.xlabel("Area (sq. ft.)", weight = 'bold')
plt.ylabel("Price (000's $)", weight = 'bold')

plt.show()
```

# Regression Scatter Plot

```python
df_ad_expenditure = pd.read_csv("scatter_plot_ii.csv")
df_ad_expenditure
```

<b>sns.regplot()<b>

```python
# plt.figure(figsize = (10,8))
sns.set(rc = {'figure.figsize':(10,8)})

sns.regplot(x = "Budget", y = "Sales",
            data = df_ad_expenditure,
            scatter_kws = {'color':'black'},
            line_kws = {'color':'red'})

plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 15, weight = 'bold')

plt.show()
```

<b>sns.lmplot()<b>

```python
sns.lmplot(x = "Budget", y = "Sales",
            data = df_ad_expenditure,
            height = 8, aspect = 1.2,
            scatter_kws = {'color':'black'},
            line_kws = {'color':'red'})

plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 15, weight = 'bold')

plt.show()
```

# Bar and Line Chart

```python
from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv("bar_line_chart_data.csv")
df_kdnuggets
```

```python
sns.set_style('white')
sns.despine()
```

```python
fig, ax = plt.subplots(figsize = (10,7))

ax.bar(df_kdnuggets['Year'],
       df_kdnuggets['Participants'],
       color = 'black')
ax.set_ylabel("Number of Participants", fontsize = 'large', weight = 'bold')
ax.tick_params(axis = 'y', width = 2, labelsize = 'large')

ax1 = ax.twinx()
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1))
ax1.plot(df_kdnuggets['Year'],
         df_kdnuggets['Python Users'],
         color = 'red',
         marker = 'D')
ax1.set_ylabel("Python Users", fontsize = 'large', weight = 'bold', color = 'red')
ax1.tick_params(axis = 'y', colors = 'red', width = 2, labelsize = 'large')

ax.set_title("KD Nuggets Survey Python Users (2012 - 2019)", fontsize = 15, weight = 'bold')
```
